# Q6. Plot a bar graph for Host_ID vs Count of Listings.

```python
idcount=ab.host_id.value_counts().head(50).sort_values(ascending=False)
idcount.plot(kind='bar',figsize=(20,10));
plt.title('Bar graph for Host_ID vs Count of Listings');
plt.xlabel('Host_ID');
plt.ylabel('Count of Listings');
```
